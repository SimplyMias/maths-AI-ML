Compute the magnitude of a vector and verify it matches the Pythagorean theorem, then modify to compute the unit vector.


```python
import jax.numpy as jnp

a = jnp.array([3.0,4.0])

magnitude = jnp.sqrt(jnp.sum(a**2))
print(f"Magnitude: {magnitude}")
```
