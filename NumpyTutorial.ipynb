**Creating Array**

```python
import numpy as np # importing numpy
```

np.zeros = it is a function which creates an array  of any number of zeros

```python
np.zeros(5)
```

To make 2D array (a matrix) provide a tuple with num of rows and num of columns
Example 3X4 matrix

```python
np.zeros((3,4))
```

**Some Vocabulary**

In Numpy each Dimension is called axis
the number of axes is called rank

Example the above 3X4 matrix is an array of rank 2

the first axis has length 3 and the second axis has  length 4

An array's list of axes is shape of the array

Eample, above matrix's shape is (3,4). The rank is equal to the shape length

The size of an array is the total number of elements, which is the product of all axis lengths (e.g. 3*4=12)

```python
a = np.zeros((3,4))
a
```
