Check whether two vectors are parallel by testing if one is a scalar multiple of the other.


```python
import jax.numpy as jnp

a = jnp.array([2,4,6])
b = jnp.array([1,2,3])

ratios = a/b
print(f"Ratios: {ratios}")
print(f"Parallel: {jnp.allclose(ratios, ratios[0])}") #jnp.allclose(ratios, ratios[0]) this checks weather all the elements in the ratios are approximately equal to the first element
```
